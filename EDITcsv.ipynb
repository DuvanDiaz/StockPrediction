```python
bynd['Date'] = pd.to_datetime(bynd.Date)
bynd['Date'] = bynd['Date'].dt.strftime('%Y/%m/%d')
bynd.drop(columns= ['Unnamed: 0'])
print(bynd)
```

```python
# Roku Stock 
roku = pd.read_csv("Datasets/ROKU.csv")
roku
```

```python
# Tesla Stock
tsla = pd.read_csv("Datasets/TSLA.csv")
tsla
```

```python
# Vanguard ETF Stock
etf = pd.read_csv("Datasets/VNFTF.csv")
etf
```

```python
# Merge three of the files into a new .csv by Symbol column. 
stockxx = pd.concat([etf, tsla, roku], sort=False)
stockxx.head(300)
```

```python
stockxx.columns=stockxx.columns.str.replace('/','')
stockxx.columns=stockxx.columns.str.replace('CloseLast','Close')
stockxx[" Close"]= stockxx[' Close'].str.replace('$', '')
stockxx[" Open"]= stockxx[' Open'].str.replace('$', '')
stockxx[" High"]= stockxx[' High'].str.replace('$', '')
stockxx[" Low"]= stockxx[' Low'].str.replace('$', '')
stockxx
```

```python
stockxx['Date'] = pd.to_datetime(bynd.Date)
stockxx['Date'] = stockxx['Date'].dt.strftime('%Y/%m/%d')
# stockxx.drop(columns= ['Unnamed: 0'])
print(stockxx)
```

```python
# Export Csv
stockxx.to_csv("Datasets/multiplestock.csv")
```
